## Existen dos iteraciones principales: For and While

## For loop

```python
for element in list_1:
  print(element)
```

```python
a = 0
for element in list_1:
  # += lo que hace es que iguala y luego suma al valor anterior
  a += element
  print(a)
```

```python
b = 1
# Definición de factorial con list_1
for f in list_1:
  b *= f
print(b)
```

```python
# Lista definida con rangos (x,y,z) x = límite inferior, y = límite superior, z = step
# El * es un argument unpacking operator (dame todos los elementos de la función range)
list_3 = [*range(1,11,1)]
list_3
```

```python
list_p = [*range(1,10,1), *range(1,20,2)]
print(list_p)
# Nótese la diferencia al poner el * y cuando no se ponde

list_q = [range(1,10,1), range(1,20,2)]
print(list_q)
```

```python
# Ordenar las listas de forma ascendente
list_p.sort()
list_p
```

```python
# Orden en orden descendente
list_p.sort(reverse=True)
list_p
```

```python
# Otra forma de hacer sort, sin embargo, no se guarda el sort si se usa esta forma
print(sorted(list_p))
```

```python
factorials = []
a = 1
for i in list_3:
  a = a * i
  # Append es el operador que agrega los elementos al final de la
  factorials.append(a)
```

```python
factorials
```

## While loop
### Cuidado con entrar en loops infinitos

¡Ojo con las convergencias y divergencias!

```python
a = 0
while a < 100:
  a += 1
  print(a)
```

```python
# Paradoja de Zenón
z = 0
for i in range(0,100,1):
    z += 1/(2**i)
    print(z)
z
```

```python
z = [0,1]
a = 1/2
while z[-1]-z[-2]>0:
    a *= 1/2
    z.append(a)
z
```

## Condicionales if

```python
# Se enuncia el condicional y luego se le dan las instrucciones de qué hacer (pueden ser múltiples instrucciones).

a = 1
if 1 < 2:
  print('True')
  print(a)
  a += 1
  print(a)
```

```python
# if luego else (si no es cierto)
a = 1
if a > 2:
  print(a)
else:
  a += 1
  print(a)
```

```python
# elif es un condicional adicional

if a<2:
  print(a)
elif a>2:
  print('3')
else:
  a+=1
  print(a)
```

```python
# Nos permite protegernos de los errores (p.e. dividir entre cero)

try:
  8/0
# El error lo conocimos al correr el códio 'ZeroDivisionError' fue lo que nos lanzó
except ZeroDivisionError:
  print('No se puede dividir entre cero')
```

```python
# Otra forma de hacerlo, sin necesidad de conocer el error

try:
  8/0
except:
  print('Hay algo mal')
```

# Collatz Conjecture

```python
n = int(input())
collatz = list()

while 1:
  if n%2 == 0:
    n = n/2
    collatz.append(n)
  elif n == 1:
    break
  else:
    n = n*3+1
    collatz.append(n)

print(collatz)
```

```python
n = int(input())
collatz = list()

while n!=1:
  if n%2 == 0:
    n = n/2
    collatz.append(n)
  else:
    n = n*3+1
    collatz.append(n)

print(collatz[-3:])
```
