## Tipos de variables y sus operaciones

```python
# Declaración de las variables en strings
a = 'Hola mundo'
c = 'Hola'
```

```python
# Declaración de varaibles tipo int o floats
b = 1
b + 3
b * 6
b ** 3
b % 3
```

### Existen 3 tipos de variable
- String
- Integer
- Float

```python
# a+b no se puede
# Tenemos que usar el comando format 'f'
d = f'{a} {b}'
d
```

```python
# Cuando el string sea mayor de 1 línea, hay que usar triple comilla
f = '''Hola mundo
Hola mundo
Hola mundo'''
f
```

```python
# Todo string es una lista, por lo tanto, podemos medirlo
len(d)
```

```python
# Para extraer un elemento de la lista usamos lo siguiente
# Ojo: las listas empiezan en 0
d[3]
```

```python
# Aunque len(d) sea 12, el último valor es 11
d[11]
```

```python
# Otra forma para tomar el último elemento
d[-1]
```

```python
# Como cambio de formato a los números
int(d[11])
```

```python
# También se pueden hacer float
float(d[11])
```

```python
# Como identificcar el tipo de variable
type(b)
```

```python
# Ejercicio: crear 2 strings y concatenar
S_1 = 'Primer mensaje'
S_2 = 'Segundo mensaje'
f' {S_1} {S_2} '
```

```python
# Otra forma de hacerlo
' ' + S_1 + ' ' + S_2 + ' '
```

```python
# Otra forma de hacerlo
' '.join([a,c])
```

```python
# Concatenar en donde el que una sea un 'enter'
print('\n'.join([a,c]))
```

```python
# Otra forma es concatenar con 'tab'
print('\t'.join([a,c]))
```
