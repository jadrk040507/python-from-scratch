```python
import itertools
from itertools import product
```

```python
Moneda = {'A','S'}
Omega = set(product(Moneda, repeat = 3))
Omega
```

```python
B = {(x,y,z) for x,y,z in Omega if x=='S'}
B
```

```python
A = {(x,y,z) for x,y,z in Omega if y=='S' or z=='S' or (y=='s' and z=='S')}
E = B.intersection(A)
E
```

```python
def P(x,y):
  print(len(x)/len(y))
```

```python
P(E,B)
```
