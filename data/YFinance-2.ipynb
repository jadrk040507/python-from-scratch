```python
#importar todos los valores de las acciones seleccionadas
from datetime import date
df_tickers = pd.read_excel('/content/gdrive/MyDrive/Colab Notebooks/Python/Yfinance/ListaEmpresasMexico.xlsx',sheet_name='Tickers')
tickers = [*df_tickers['Ticker YF']]
data_frames = {}
end_date = date.today().strftime('%Y-%m-%d')
for ticker in tickers:
    data = yf.download(ticker, end=end_date)
    data_frames[ticker] = data
```

```python
data_frames['AC.MX'] #ArcaContinental
```

```python
data_frames['AC.MX'].describe()
```

```python
#Correlación todos contra todos.
import matplotlib.pyplot as plt
correlation_matrix = pd.DataFrame()
for ticker1, data1 in data_frames.items():
    for ticker2, data2 in data_frames.items():
        if ticker1 != ticker2:
            correlation = data1['Adj Close'].corr(data2['Adj Close'])
            correlation_matrix.loc[ticker1, ticker2] = correlation

plt.figure(figsize=(100, 80))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Stock Correlation Heatmap', fontsize = 14)
plt.show()
```

```python
correlation_matrix['KOF']
```

```python
#Correlación de todos los elementos con su top 10 de acciones.
correlation_matrices = {}
for stock_symbol in data_frames.keys():
    correlations = correlation_matrix[stock_symbol].drop(stock_symbol)
    top_correlated_stocks = correlations.nlargest(10).index.tolist()
    correlation_matrices[stock_symbol] = correlation_matrix[top_correlated_stocks].loc[top_correlated_stocks]
for stock_symbol, correlation_matrix in correlation_matrices.items():
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'Correlation Matrix for {stock_symbol} and Top 10 Correlated Stocks')
    plt.show()
```

```python
correlation_matrices
```

```python
#Ver las relaciones de todas las acciones como una red
import networkx as nx
for c_t in [*range(10,101,10)]:
  G = nx.Graph()
  correlation_threshold = c_t/100
  G.add_nodes_from(correlation_matrix.columns)
  for stock1 in correlation_matrix.columns:
      for stock2 in correlation_matrix.columns:
          if stock1 != stock2:
              correlation = correlation_matrix.loc[stock1, stock2]
              if correlation >= correlation_threshold:
                G.add_edge(stock1, stock2, weight=correlation)
  pos = nx.spring_layout(G)
  edge_widths = [2 * G[u][v]['weight'] for u, v in G.edges()]
  nx.draw_networkx_nodes(G, pos, node_size=200, node_color='lightblue')
  nx.draw_networkx_edges(G, pos, width=edge_widths, edge_color=edge_widths, edge_cmap=plt.cm.Blues, edge_vmin=correlation_threshold)
  labels = {node: node for node in G.nodes()}
  nx.draw_networkx_labels(G, pos, labels, font_size=8)
  plt.title(f'Stocks with Strong Correlations Graph with a Correlation Threshold of: {correlation_threshold}')
  plt.axis('off')
  plt.show()
```
