```python
import random
import statistics as stats

def sntptsb(iter=1000000):
  final_rich = []
  counter = 0
  while counter < iter:
    turns = 0
    wager = 2
    rand_i = 1
    while rand_i != 0:
      rand_i = random.randint(0, 1)
      if rand_i == 1:
        turns +=1
    total = wager**turns
    final_rich.append(total)
    counter +=1
  return print('The geometric mean is',stats.geometric_mean(final_rich))#, final_rich
```

```python
sntptsb()
```

```python
# Podemos definir un valor que al llamar la función sólo necesitamos un valor.

def nom_func(rest, iterations=0):
  if iterations == 0:
    iterations += 1
  return iterations + rest

nom_func(2)
```

```python
# Podemos definir un valor que al llamar la función sólo necesitamos un valor.

def nom_func(rest, iterations=0, *iterations_1):
  if iterations == 0:
    iterations += 1
  return iterations + rest, iterations-rest, iterations_1
```

```python
nom_func(2)
```

# Juego fizzbuzz



```python
def fizzbuzz(div1, div2, ranged):
  fb = []
  for i in range(1,ranged+1):
    if i % div1 == 0 and i % div2 == 0:
      i = 'FizzBuzz'
      fb.append(i)
    elif i % div1 == 0 and i % div2 != 0:
      i = 'Fizz'
      fb.append(i)
    elif i % div1 != 0 and i % div2 == 0:
      i = 'Buzz'
      fb.append(i)
    else:
      fb.append(i)
  return fb
```

```python
fizzbuzz(3,5,100)
```

```python
# BuzzFizz sin usar módulo
def fizzbuzz_no_module(div1, div2, ranged):
  fb = []
  for i in range(ranged+1):
    if i/div1 - int(i/div1) == 0 and i/div2 - int(i/div2) == 0:
      i = 'FizzBuzz'
      fb.append(i)
    elif i/div1 - int(i/div1) == 0:
      i = 'Fizz'
      fb.append(i)
    elif i/div2 - int(i/div2) == 0:
      i = 'Buzz'
      fb.append(i)
    else:
      fb.append(i)
  return fb
```

```python
fizzbuzz_no_module(1,6,100)
```

# Funciones lambda

Normalmente son funciones que tienen *un solo paso*.

```python
lbm_1 = lambda x,y : x+y
```

```python
lbm_1(2,3)
```

```python
lbm_1 = lambda x,y : True if x>y else False
```

```python
lbm_1(3,5)
```

```python
buzzfizz_lambda = lambda div1, div2, ranged, i=1: ['BuzzFizz' if i % div1 == 0 and i % div2 == 0 else 'Buzz' if i % div1 == 0 else 'Fizz'  if i % div2 == 0 else i for i in range(i,ranged+1)]
```

```python
buzzfizz_lambda(27,73,10)
```

```python
print(*map(lambda i: 'Fizz'*(not i%3)+'Buzz'*(not i%5), range(1,101)),sep='\n')
```
