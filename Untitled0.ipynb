```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Lee el primer archivo CSV con los datos de tenencia de bonos M por extranjeros
df1 = pd.read_csv('/content/drive/MyDrive/bonos.csv', parse_dates=True, na_values=['N/E',0], header=0, names=['Fecha', 'Tenencia Bonos M'],).dropna()

# Lee el segundo archivo CSV con los datos del cierre del tipo de cambio diario
df2 = pd.read_csv('/content/drive/MyDrive/fix.csv', parse_dates=True, na_values=['N/E',0], header=0, names=['Fecha', 'FIX']).dropna()

# Asegúrate de que las columnas de las fechas estén en formato de fecha
df1['Fecha'] = pd.to_datetime(df1['Fecha'])
df2['Fecha'] = pd.to_datetime(df2['Fecha'])

# Extrae el año y el mes de las fechas para ambos DataFrames
df1['Año'] = df1['Fecha'].dt.year
df1['Mes'] = df1['Fecha'].dt.month
df2['Año'] = df2['Fecha'].dt.year
df2['Mes'] = df2['Fecha'].dt.month

# Agrupa y promedia los datos según el mismo mes del mismo año


df1_promedio = df1.groupby(['Año', 'Mes']).mean().reset_index()
df2_promedio = df2.groupby(['Año', 'Mes']).mean().reset_index()

# Fusiona los DataFrames promediados utilizando la columna de meses y años
df_final = pd.merge(df1_promedio, df2_promedio, on=['Año', 'Mes'], suffixes=('_bonos', '_tipo_cambio'))

# Muestra el DataFrame final con los datos combinados y promediados
print(df_final)
```

```python
df_final
```

```python
df_final_con_bonos = pd.merge(df_final, df1, on=['Año', 'Mes'])
```

```python
df_final_con_bonos.dropna(inplace=True)
```

```python
df_final_con_bonos.to_csv('percepcion_riesgo.csv')
```

```python
# Crear una figura y un primer eje
fig, ax1 = plt.subplots()

# Trazar los datos de bonos en el primer eje (izquierda)
sns.lineplot(data=df_final_con_bonos, x='Mes', y='Tenencia Bonos M', ax=ax1, color='b')
ax1.set_ylabel('Datos de bonos', color='b')

# Crear un segundo eje y configurarlo para que comparta el mismo eje x
ax2 = ax1.twinx()

# Trazar los datos del tipo de cambio en el segundo eje (derecha)
sns.lineplot(data=df_final_con_bonos, x='Mes', y='FIX', ax=ax2, color='r')
ax2.set_ylabel('Tipo de cambio', color='r')

# Añadir etiquetas y título
plt.xlabel('Mes')
plt.title('Datos de bonos y tipo de cambio')

# Mostrar el gráfico
plt.show()
```
