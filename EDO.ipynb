## Ecuación de Bernoulli

\begin{align*}
xy' + y = y^2 \ln(x) \\
xy^{-2}y' + y^{-1} = \ln(x) \\
z = y^{-1} \\
z' = -y^{-2}y' \\
-xz' + z = \ln(x) \\
z' + \frac{-1}{x}z = \frac{\ln(x)}{x} \\
\mu(x) = e^{\int \frac{-1}{x} dx} = \frac{1}{x} \\
\int \frac{d}{dx} \frac{1}{x}z dx = \int \frac{\ln(x)}{x^2} dx \\
\frac{1}{x}z = \frac{-\ln(x)}{x}-\frac{1}{x} \\
z = -\ln(x) - 1 \\
y = (-\ln(x) - 1)^{-1}
\end{align*}

## (3) Ecuación de Bernoulli

\begin{align*}
y' + 2xy = 2xy^2 \\
y^{-2}y' + 2xy^{-1} = 2x \\
z = y^{-1} \\
z' = -y^{-2}y' \\
-z' + 2xz = 2x \\
z' - 2xz = -2x \\
\mu(x) = e^{\int -2x dx} = e^{-x^2} \\

z = 1 + C e^{x^2} \\
y = (1 + C e^{x^2})^{-1}
\end{align*}
