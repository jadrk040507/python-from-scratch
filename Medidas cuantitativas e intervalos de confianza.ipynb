# Ejercicio 1

El estimador de máxima verosimilitud (MLE por sus siglas en inglés) es un método para estimar los parámetros de una distribución de probabildidad, maximizando la versolimilitud de observar los datos de una muestra. En otras palabras encuentra los valores de los parámetros que hacen que los datos sean más probables bajo la distribución propuesta.

Aplicaciones en la economía.


1. Se utiliza para ajustar modelos ARIMA (Autoregressive Integrated Moving Average) y modelos de series temporales estacionales, que son funamentales para predecir y entender la dinámica de variables como el PIB, el desempleo y la inflación.

2. Se utiliza para ajustar modelos de elección discreta, como los modelos de elección logit y probit. Estos modelos se utilizan para eanalizar decisiones de consumo, preferencias del consumidor y comportamiento de compra.

3.  Se utiliza para estimar los parámetros de algunas funciones de producción como la Cobb-Douglas, que relaciona la cantidad de producción con los insumos de capital y trabajo.

4. Se utiliza para estimar los parámetros de los modelos de oferta y demanda que describen el comportamiento entre precios y cantidades.

5. Se utiliza para ajustar modelos de distribución de rendimientos, como la distribución normal, la t de Student y la distribución de la cola pesada, que son fundamentales para evaluar y gestionar el riesgo financiero.

# Ejercicio 2

Sea $X_1, X_2, \ldots, X_n$ una muestra aleatoria con distribución

$$ f_X(x) = \theta \left( x - \frac{1}{2}\right) + 1 $$
$$\forall x \in [0,1]$$

donde $\theta \in [-2, 2]$ es un parámetro desconocido. Defina

$$ \hat{\Theta}_n = 12 \bar{X}_n -6$$
para estimar a $\theta$.

## ¿Es $\hat{\Theta}_n$ insesgado?

\begin{align}
\mathbb{E} \left[ X \right] &= \int_0^1 x \left(\theta \left( x - \frac{1}{2}\right) + 1 \right) dx \\
\mathbb{E} \left[ X \right] &= \int_0^1 \theta x^2 - \theta x\frac{1}{2} + x dx \\
\mathbb{E} \left[ X \right] &= \left. \theta \frac{x^3}{3} - \theta \frac{x^2}{4} + \frac{x^2}{2} \right|_0^1 \\
\mathbb{E} \left[ X \right] &= \frac{\theta}{3} -  \frac{\theta}{4} + \frac{1}{2} \\
\mathbb{E} \left[ X \right] &= \frac{\theta}{12} + \frac{1}{2} \\
\end{align}

\begin{align}
\mathbb{E} \left[ \hat{\Theta}_n \right] &= 12 \mathbb{E}[\bar{X}_n] - 6 \\
\mathbb{E} \left[ \hat{\Theta}_n \right] &= \frac{12}{n} n\mathbb{E}[X] - 6 \\
\mathbb{E} \left[ \hat{\Theta}_n \right] &= 12 \left( \frac{\theta}{12} + \frac{1}{2} \right) - 6 \\
\mathbb{E} \left[ \hat{\Theta}_n \right] &= \theta + 6 - 6 \\
\mathbb{E} \left[ \hat{\Theta}_n \right] &= \theta \\
\end{align}

\begin{align}
Bias(\hat{\Theta}_n) = \mathbb{E} \left[ \hat{\Theta}_n \right] - \theta \\
Bias(\hat{\Theta}_n) = \theta - \theta = 0
\end{align}

Por lo tanto, el estimador es insesgado.

## Calcula $MSE(\hat{\Theta}_n)$

\begin{align}
\mathbb{E} \left[ X^2 \right] &= \int_0^1 x^2 \left(\theta \left( x - \frac{1}{2}\right) + 1 \right) dx \\
\mathbb{E} \left[ X^2 \right] &= \int_0^1 \theta x^3 - \theta x^2\frac{1}{2} + x^2 dx \\
\mathbb{E} \left[ X^2 \right] &= \left. \theta \frac{x^4}{4} - \theta \frac{x^3}{6} + \frac{x^3}{3} \right|_0^1 \\
\mathbb{E} \left[ X^2 \right] &= \frac{\theta}{4} -  \frac{\theta}{6} + \frac{1}{3} \\
\mathbb{E} \left[ X^2 \right] &= \frac{\theta}{12} + \frac{1}{3} \\
\end{align}

\begin{align}
Var(X) &= \frac{\theta}{12} + \frac{1}{3} - \left( \frac{\theta}{12} + \frac{1}{2} \right)^2\\
Var(X) &= \frac{\theta}{12} + \frac{1}{3} -  \frac{\theta^2}{144} - \frac{\theta}{12} - \frac{1}{4}\\
Var(X) &= \frac{1}{3} -  \frac{\theta^2}{144} - \frac{1}{4}\\
Var(X) &=  -  \frac{\theta^2}{144} + \frac{1}{12}\\
\end{align}

\begin{align}
MSE(\hat{\Theta}_n) &= Var(\hat{\Theta}_n) + Bias(\hat{\Theta}_n) \\
MSE(\hat{\Theta}_n) &= Var(\hat{\Theta}_n) \\
MSE(\hat{\Theta}_n) &= Var(12 \bar{X}_n - 6) \\
MSE(\hat{\Theta}_n) &= \frac{144}{n^2}nVar(X) \\
MSE(\hat{\Theta}_n) &= \frac{144}{n} \left( -  \frac{\theta^2}{144} + \frac{1}{12} \right) \\
MSE(\hat{\Theta}_n) &= -  \frac{\theta^2}{12} + \frac{12}{n}
\end{align}

# Ejercicio 3

El numero de tiros de tres durante un partido de la NBA se distribuye como una variable aleatoria Poisson. El domingo, el numero de tiros de tres logrados fueron 14, 26, 25 y 13.

Calcula $\hat{\lambda}_{MLE}$.

\begin{align}
L(x_1, x_2, \ldots, x_n) &= \prod_{i=1}^n \frac{\lambda^{x_i}}{x_i!} e^{-\lambda} \\
l(x_1, x_2, \ldots, x_n) &= \sum_{i=1}^n \left( x_i\ln(\lambda) - \ln(x_i!) -\lambda \right) \\
\frac{d}{d\lambda} l(x_1, x_2, \ldots, x_n) &= \sum_{i=1}^n \left( \frac{x_i}{\lambda} - 1 \right) = 0 \\
0 &= \frac{1}{\lambda} \sum_{i=1}^n x_i - n \\
\frac{n}{\sum_{i=1}^n x_i} &= \frac{1}{\lambda} \\
\hat{\lambda}_{MLE} &= \bar{X}_n
\end{align}

# Ejercicio 6

Sea $X_1, \ldots, X_n$ una muestra aleatoria que se tomaa partir de una distribución con varianza 16. Encuentra un intervalo de confianza al nivel $0.05$ para $\theta = \mathbb{E}[X_i]$.

```python
import numpy as np
import scipy.stats as st
sample = np.random.normal(scale = 4, size = 100)
```

```python
# T-test
st.t.interval(confidence=0.95, df=len(sample)-1, loc=np.mean(sample), scale=16**(1/2))
```

```python
# Z value
z = st.norm.ppf(1-0.05/2)
(np.mean(sample) - z*np.std(sample), np.mean(sample) + z*np.std(sample))
```

# Ejercicio 7

To estimate the portion of voters who plan to vote for Candidate A in an election, a random sample of size $n$ from the voters is chosen. The sampling is done with replacement. Let $\theta$ be the portion of voters who plan to vote for Candidate A among all voters. How large does $n$ need to be so that we can obtain a $90\%$ confidence interval with $3\%$ margin of error? That is, how large $n$ needs to be such that

$$\mathbb{P} \left( \bar{X} - 0.03 \leq \theta \leq \bar{X} + 0.03 \right) \geq 0.90 $$

Asumimos que $X_i \sim Bin{\theta}$

Por lo tanto:

\begin{align}
\max_\theta Var(X) &= \theta(1-\theta) \\
1 - 2\theta &= 0 \\
\theta_\max &= \frac{1}{2}
\end{align}

Entonces, el intervalo de confianza con una varianza máxima sería:

$$ \left[ \bar{X} - \frac{z_{\alpha/2}}{2 \sqrt{n}}, \bar{X} + \frac{z_{\alpha/2}}{2 \sqrt{n}} \right] $$

Dado que $\alpha = 0.10$, $z_{\alpha/2} = 1.645$.

Nos quedaría entonces que:

\begin{align}
\frac{z_{\alpha/2}}{2 \sqrt{n}} &= 0.03 \\
\frac{1.645}{2 \sqrt{n}} &= 0.03 \\
\frac{1}{\sqrt{n}} &= \frac{0.03*2}{1.645} \\
\sqrt{n} &= \frac{1.645}{0.03*2} \\
n &\approx 752
\end{align}
